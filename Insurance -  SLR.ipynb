```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
data = pd.read_csv("insurance.csv")
data.head()
```

### Age vs Insurance Charges Linear Regression Model

```python
# Selecting all rows from columns 'age' and 'charges' 
x1 = data.iloc[:,:1].values
y1 = data.iloc[:,[6]].values

# Scatter plot of ID vs DV
plt.scatter(x1,y1,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fitting a model to the data concerning 'age' and 'charges'
age_insurance_model = LinearRegression()
age_insurance_model.fit(x1,y1)
```

```python
# Creating model for predicitions using IV values and the fitting model
y_pred1 = age_insurance_model.predict(x1)

# Scatter plot of IV vs DV with line of best fit 
plt.scatter(x1,y1,color = 'b')
plt.plot(x1,y_pred1,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

#### Example of a prediction for insurance for a person who is 26 years old.

```python
# Selecting an IV value for 'age' where the DV value 'charge' is unknown 
unk_x1 = [[26]]

# Reshaping the our model based on the number of samples and number of features
x1_pred = np.append(x1, unk_x1).reshape(-1,1)

# Using the reshaped values to make prediciton for the DV
y1_pred = age_insurance_model.predict(x1_pred)

plt.scatter(x1,y1,color = 'b')
plt.plot(x1_pred,y1_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print(unk_x1,"year old would be charged approximately", age_insurance_model.predict(unk_x1), "for insurance.")
```

#### Example of a prediction for insurance for a person who is 55 years old.

```python
# Selecting an IV value for 'age' where the DV value 'charge' is unknown 
unk_x2 = [[55]]

plt.scatter(x1,y1,color = 'b')
plt.plot(x1_pred,y1_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print(unk_x2,"year old would be charged approximately", age_insurance_model.predict(unk_x2), "for insurance.")
```

It appears that the older the individual the higher their insurance.

The simple linear regression models appears to be estimating values correctly, thus it is an appropriate model to use for this dataset to estimate insurance charges based on an individuals age.

#### Example of a prediction for insurance for a person who's BMI is 32.
